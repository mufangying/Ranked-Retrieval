# Visualizations: varying $\beta$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.load('varying_beta_yahoo.npy')
results.shape
```

## trajectory visualization

```python
colors = ['red', 'blue', 'black']
labels = ['tCRC', 'tCRC-s', 'LTT']
markers = ['s', 'o', '^']

plt.figure()
plt.xlim(0.960,1.001)
plt.ylim(0.970,1.001)

for i in range(3):
    x = results[:,i,0]
    y = results[:,i,1]
    plt.scatter(x, y, color=colors[i], label= labels[i], marker = markers[i])

    # Plot lines connecting the points
    plt.plot(x, y, color=colors[i] )
```

```python
colors = ['red', 'blue', 'black']
labels = ['tCRC', 'tCRC-s', 'LTT']
markers = ['s', 'o', '^']
response_names = [r'$\hat{\lambda}$', r'$\hat{\gamma}$', 'Risk (1)', 'Risk (2)', 
                  'Prediction set size', r'Recall $(\geq 2)$', r'Recall $(1)$', 'Precision']

beta_seq = [0.05, 0.075, 0.1, 0.125, 0.15, 0.175, 0.2]
```

## Prediction set size 

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 4 rows, 2 columns

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create 8 different plots
for i, ax in enumerate(axes):
    ax.scatter(np.array(beta_seq),  results[:,0,i], c = colors[0], label= labels[0], marker = markers[0])
    ax.plot(np.array(beta_seq),  results[:,0,i], c = colors[0])
    
    ax.scatter(np.array(beta_seq),  results[:,1,i], c = colors[1], label= labels[1], marker = markers[1])
    ax.plot(np.array(beta_seq),  results[:,1,i], c = colors[1])
    
    ax.scatter(np.array(beta_seq),  results[:,2,i], c = colors[2], label= labels[2],marker = markers[2])
    ax.plot(np.array(beta_seq),  results[:,2,i], c = colors[2])
    
    ax.set_xlabel(r'$\alpha_2$', fontsize = 15)
    ax.set_ylabel(response_names[i], fontsize = 15)
    ax.set_title(f'{response_names[i]} vs. '+ r'$\alpha_2$', fontsize = 18, y = 1.01)
    ax.legend(fontsize= 16)
    ax.grid(True)

# Adjust layout for better spacing
plt.tight_layout()

# Show the figure
plt.savefig('yahoo_parameters.pdf')
```
